```python
import pandas as pd

# === 1. 读取文件 ===
f1 = pd.read_csv("feature_importance_original_order(1).csv")
f2 = pd.read_csv("user_asset_preferences_full.csv", nrows=1)  # 只读表头就够

# === 2. 提取第二份文件中“第二列开始”的列名 ===
ordered_features = f2.columns[1:].tolist()

# === 3. 按这个顺序重新排列 f1 的行（整行移动） ===
f1_sorted = f1.set_index("feature").loc[ordered_features].reset_index()

# === 4. 导出结果 ===
f1_sorted.to_csv("feature_importance_reordered.csv", index=False)

print("✅ 已按 user_asset_preferences_full 的列顺序重新排列行！输出：feature_importance_reordered.csv")
```
